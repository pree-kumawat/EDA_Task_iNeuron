```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
url = 'https://raw.githubusercontent.com/pree-kumawat/EDA_Task_iNeuron/main/Census%20Dataset.CSV'
df = pd.read_csv(url)
```

```python
df
```

```python
df.head()
```

```python
# Checking the shape of the dataset
print('Number of rows:', df.shape[0])
print('Number of columns:', df.shape[1])
```

```python
df.shape
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Checking the unique values in the categorical columns
print('Workclass:', df['Workclass'].unique())
print('Education:', df['Education'].unique())
print('Marital-status:', df['Marital-status'].unique())
print('Occupation:', df['Occupation'].unique())
print('Relationship:', df['Relationship'].unique())
print('Race:', df['Race'].unique())
print('Sex:', df['Sex'].unique())
print('Native-country:', df['Native-country'].unique())
print('Salary:', df['Salary'].unique())
```

```python
# Dropping the missing values
df = df.replace(' ?', np.nan)
df = df.dropna()
```

```python
df
```

```python
# Checking the shape of the dataset after dropping the missing values
print('Number of rows:', df.shape[0])
print('Number of columns:', df.shape[1])
```

```python
df.shape
```

```python
# Distribution of age
plt.figure(figsize=(12,6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

## We can see that the majority of the individuals in the dataset are between the ages of 20 and 50. 

```python
# Distribution of workclass
plt.figure(figsize=(12,6))
sns.countplot(x='Workclass', data=df)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Distribution of Workclass')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset work in the private sector

```python
# Distribution of education
plt.figure(figsize=(12,6))
sns.countplot(x='Education', data=df)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Distribution of Education')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset have completed high school or some college.

```python
# Distribution of marital status
plt.figure(figsize=(12,6))
sns.countplot(x='Marital-status', data=df)
plt.xlabel('Marital-status')
plt.ylabel('Count')
plt.title('Distribution of Marital Status')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset are either married or never married.

```python
# Distribution of occupation
plt.figure(figsize=(12,6))
sns.countplot(x='Occupation', data=df)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Distribution of Occupation')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset work in the sales, craft-repair, or professional specialty occupations.



```python
# Distribution of relationship
plt.figure(figsize=(12,6))
sns.countplot(x='Relationship', data=df)
plt.xlabel('Relationship')
plt.ylabel('Count')
plt.title('Distribution of Relationship')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset are either husbands or not in a family.



```python
# Distribution of race
plt.figure(figsize=(12,6))
sns.countplot(x='Race', data=df)
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Distribution of Race')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset are white.



```python
# Distribution of sex
plt.figure(figsize=(12,6))
sns.countplot(x='Sex', data=df)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Distribution of Sex')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the dataset is skewed towards males.



```python
# Distribution of native country
plt.figure(figsize=(12,6))
sns.countplot(y='Native-country', data=df)
plt.xlabel('Count')
plt.ylabel('Native-country')
plt.title('Distribution of Native Country')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset are from the United States.



```python
# Distribution of salary
plt.figure(figsize=(12,6))
sns.countplot(x='Salary', data=df)
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Distribution of Salary')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the majority of the individuals in the dataset have an income less than or equal to 50k.



# Relationship between variables


```python
# Relationship between age and salary
plt.figure(figsize=(12,6))
sns.boxplot(x='Salary', y='Age', data=df)
plt.xlabel('Salary')
plt.ylabel('Age')
plt.title('Relationship between Age and Salary')
plt.xticks(rotation=45)
plt.show()
```

## We can see that the median age for individuals with an income greater than 50k is higher than the median age for individuals with an income less than or equal to 50k.



```python
# Relationship between education and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Education', hue='Salary', data=df)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Relationship between Education and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that individuals with higher education levels are more likely to have an income greater than 50k.



```python
# Relationship between occupation and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Occupation', hue='Salary', data=df)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Relationship between Occupation and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that individuals in certain occupations (such as executive managerial, professional specialty, and protective service) are more likely to have an income greater than 50k.



```python
# Relationship between sex and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Sex', hue='Salary', data=df)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Relationship between Sex and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that males are more likely to have an income greater than 50k than females.



```python
# Relationship between race and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Race', hue='Salary', data=df)
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Relationship between Race and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that individuals of the White race are more likely to have an income greater than 50k than individuals of other races.



```python
# Relationship between workclass and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Workclass', hue='Salary', data=df)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Relationship between Workclass and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that individuals who work in the private sector are more likely to have an income greater than 50k than individuals who work in other workclasses.



```python
# Relationship between marital status and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Marital-status', hue='Salary', data=df)
plt.xlabel('Marital-status')
plt.ylabel('Count')
plt.title('Relationship between Marital-status and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

## We can see that individuals who are married are more likely to have an income greater than 50k than individuals who are not married.



```python
# Relationship between hours per week and salary
plt.figure(figsize=(12,6))
sns.boxplot(x='Salary', y='Hours-per-week', data=df)
plt.xlabel('Salary')
plt.ylabel('Hours-per-week')
plt.title('Relationship between Hours-per-week and Salary')
plt.xticks(rotation=45)
plt.show()
```

## We can see that individuals who have an income greater than 50k tend to work more hours per week than individuals who have an income less than or equal to 50k.

We discovered that the majority of the individuals in the dataset are from the United States and have an income less than or equal to 50k.
